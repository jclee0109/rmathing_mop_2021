```python
import sqlite3
import pandas as pd
import numpy as np

# Create the connection
db = sqlite3.connect(r'C:\Users\이주찬\Desktop\rmathing\mop_rm\Giganpyo_s\timesite\db.sqlite3')

# create the dataframe from a query
df = pd.read_sql_query("SELECT * FROM timetable_subjectinfo", db)
df
```

```python
subject = pd.read_sql_query("SELECT * FROM timetable_subjectinfo", db)
evaluation =pd.read_sql_query("SELECT * FROM timetable_evaluation", db)
```

```python
evaluation
```

```python
evaluation['test'] = np.array([3,3,4,4.5], dtype = 'float64')
evaluation['assignment'] = np.array([2.5,3,3,4], dtype = 'float64')
evaluation['grade'] = np.array([3,4,4,4.5], dtype = 'float64')
```

```python
evaluation = evaluation.append({'id': 5, 'test':3.5, 'assignment':4.5, 'grade':3.5, 'comment':'sample1 좋아요~', 'subject_id':176}, ignore_index=True)
evaluation = evaluation.append({'id': 6, 'test':2.5, 'assignment':4.0, 'grade':3.0, 'comment':'sample2 좋아요~', 'subject_id':176}, ignore_index=True)
evaluation = evaluation.append({'id': 7, 'test':4.5, 'assignment':4.0, 'grade':3.0, 'comment':'sample3 좋아요~', 'subject_id':176}, ignore_index=True)
evaluation = evaluation.append({'id': 8, 'test':2.0, 'assignment':3.5, 'grade':4.0, 'comment':'sample4 좋아요~', 'subject_id':177}, ignore_index=True)
evaluation = evaluation.append({'id': 9, 'test':3.0, 'assignment':3.0, 'grade':4.5, 'comment':'sample5 좋아요~', 'subject_id':178}, ignore_index=True)
evaluation
```

```python
subject_eval = subject.merge(evaluation, left_on = 'id', right_on = 'subject_id').loc[:, ['name', 'professor1', 'professor2', 'year', 'session', 'test', 'assignment', 'grade', 'comment', 'subject_id']]
```

```python
subject_eval
```

```python
subject_eval[(subject_eval['name'] == '분자생물학') & (subject_eval['professor1'] == '다런윌리엄스')]
mean_test = np.mean(subject_eval[(subject_eval['name'] == '분자생물학') & (subject_eval['professor1'] == '다런윌리엄스')]['test'])
mean_assignment = np.mean(subject_eval[(subject_eval['name'] == '분자생물학') & (subject_eval['professor1'] == '다런윌리엄스')]['assignment'])
mean_grade = np.mean(subject_eval[(subject_eval['name'] == '분자생물학') & (subject_eval['professor1'] == '다런윌리엄스')]['grade'])

mean_test, mean_assignment, mean_grade
```

```python
subject_eval = subject_eval.groupby(['name', 'professor1']).mean().reset_index()
subject_eval
```
